## Load IMDB Dataset through keras

1. load imdb dataset with keras dataset API
2. sort datset by review length
3. save data by pickle


```python
import pickle
```

```python
from tensorflow.keras.datasets import imdb
```

```python
kargs = {'start_char' : None,
         'oov_char' : 1,
         'index_from' : 1}
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(**kargs)
```

```python
word_index = imdb.get_word_index()
```

```python
train_x, train_y = map(list, zip(*sorted(zip(x_train, y_train), key = lambda x: len(x[0]))))
```

```python
print([len(x) for x in train_x[:30]])
```

```python
test_x, test_y = map(list, zip(*sorted(zip(x_test, y_test), key = lambda x: len(x[0]))))
```

```python
IMDB_PATH = './data/imdb/imdb.data'
IMDB_VOCAB_PATH = './data/imdb/imdb.voc'
```

```python
imdb_data = ((train_x, train_y), (test_x, test_y))
```

```python
pickle.dump(imdb_data, open(IMDB_PATH, 'wb'))
pickle.dump(word_index, open(IMDB_VOCAB_PATH, 'wb'))
```
